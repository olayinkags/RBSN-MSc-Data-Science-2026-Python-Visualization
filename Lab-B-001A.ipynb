```python
# Import necessary libraries
import matplotlib.pyplot as plt
```

```python
# Define the data
x = ['Imo', 'Abuja', "Kano", 'Lagos', "Oyo"]
y = [22, 30, 45, 74, 11]
```

```python
# Create a bar chart
fig, ax = plt.subplots()
ax.bar(x, y, color='orange')

# Add title and labels
ax.set_title('Sample Bar Chart', fontsize=16, fontweight='bold', color='navy')
ax.set_xlabel('X-axis', fontsize=10, fontweight='bold', color='navy')
ax.set_ylabel('Y-axis', fontsize=10, fontweight='bold', color='navy')

# Handle grid display
ax.grid(False)

# Set figure size
fig.set_size_inches(8, 6)

# Show the plot
plt.show()
```
