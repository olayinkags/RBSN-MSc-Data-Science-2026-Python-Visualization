```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(7, 7))

fruits = {'Apple': 450, 'Banana': 200, 'Cherry': 780, 'Date': 850}
values = list(fruits.values())

# produce a pie chart
ax.pie(
    values,
    labels=list(fruits.keys()),
    labeldistance=1.05,
    startangle=90,
    autopct="%.0f%%",
    pctdistance=0.75,
    textprops={"color": "white", "weight": "normal", "size": 12},
)

ax.legend(list(fruits.keys()), loc="best")

# create a hole that goes into the centre of the pie chart
hole = plt.Circle((0,0), 0.50, color="white")
plt.gcf()

# add the hole to the pie chart
fig.gca().add_artist(hole)
```
