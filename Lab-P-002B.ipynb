```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(7, 7))

fruits = {'Apple': 450, 'Banana': 200, 'Cherry': 780, 'Date': 850}

# produce a pie chart using the values from the dictionary
ax.pie(
    # All the rows must be numeric
    list(fruits.values()),

    # Labels for each segment
    labels=list(fruits.keys()),

    # Sets start angle for the pie chart
    startangle = 90,

    # formats the display using the actual values instead of percentages
    autopct=lambda pct: f"{int(round(pct / 100.0 * sum(fruits.values())))}",

    # defines colours for the different segments
    colors=["green", "red", "orange", "grey"],

    # defines which segment blows out based on the number of segments
    explode=[0, 0, 0, 0.1],

    # creates a shadow effect
    shadow=True
    )
```
