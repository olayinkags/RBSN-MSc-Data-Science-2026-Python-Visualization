```python
# Import necessary libraries
import matplotlib.pyplot as plt
```

```python
# Define the data
x = [1, 2, 3, 4, 5]
y = [22, 30, 45, 74, 11]
```

```python
# Create a line plot
fig, ax = plt.subplots()
ax.plot(x, y, color='g', linestyle='dotted', linewidth=2, marker='o', markersize=8, markerfacecolor='red', markeredgewidth=2, markeredgecolor='black')

# Color single letter options: 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'
# Color names: 'blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'white', etc.

# Add title and labels
ax.set_title('Sample Line Plot', fontsize=16, fontweight='bold', color='blue')
ax.set_xlabel('X-axis', fontsize=16, fontweight='bold', color='blue')
ax.set_ylabel('Y-axis', fontsize=16, fontweight='bold', color='blue')

ax.set_ylim(0, 100)  # Set y-axis limits
ax.set_xlim(0, 6)     # Set x-axis limits

# Handle grid display
ax.grid(False)

# Add annotation
ax.annotate('Peak Value', xy=(4, 74), xytext=(3, 80), arrowprops=dict(facecolor='black', shrink=0.05))

# Add horizontal line at y=50
ax.axhline(y=50, color='red', linestyle='--', linewidth=1)
ax.text(0.5, 50, 'Horizontal Line', color='red', fontsize=12, verticalalignment='center')

# Set figure size
fig.set_size_inches(8, 6)

# Show the plot
plt.show()
```
