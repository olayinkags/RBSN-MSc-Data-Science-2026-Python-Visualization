```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Sales data
salesData = {
    "SalesYear": [2018, 2019, 2020, 2021, 2022],
    "Revenue": [150, 200, 250, 300, 350],
    "Profit": [50, 80, 120, 150, 200],
    "Expenses": [100, 120, 130, 150, 170],
}
```

```python
# Generate a dataframe using the data
df = pd.DataFrame(salesData)

df.head()
```

```python
# Create a hybrid chart with revenue in bar chart and profit in line chart
fig, ax1 = plt.subplots(figsize=(12, 8))

ax2 = ax1.twinx()

bars = ax1.bar(df.SalesYear, df.Revenue, color="#1f77b4", alpha=0.6, label="Revenue")
ax2.plot(df.SalesYear, df.Profit, color="#ff7f0e", marker="o", markersize=8, linestyle='-.', linewidth=2, label="Profit")

# add revenue labels inside the center of each bar
for bar in bars:
    height = bar.get_height()
    ax1.text(
        bar.get_x() + bar.get_width() / 2,
        height / 2,
        f"${height}",
        ha='center',
        va='center',
        color='white',
        fontsize=10,
        fontweight='bold'
    )

# add profit labels above each marker
for x, y in zip(df.SalesYear, df.Profit):
    ax2.text(
        x,
        y + 5,
        f"${y}",
        ha='center',
        va='bottom',
        color='#ff7f0e',
        fontsize=10,
        fontweight='bold'
    )

ax1.set_xlabel("Sales Year")
ax1.set_ylabel("Revenue", color="#1f77b4")
ax2.set_ylabel("Profit", color="#ff7f0e")

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left')

plt.title("Revenue and Profit Over Years")

plt.show()
```
