```python
# Bar Chart demonstration in Matplotlib
# Horizontal Bar Chart with different colors for each bar and data labels on top of each bar.
# Date: Saturday, 9th May 2026
# Author: Ibrahim Hammed
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
```

```python
# Define the data
x = ['Imo', 'Abuja', "Kano", 'Lagos', "Oyo"]
y = [22, 30, 45, 74, 11]
stateColor = ['red', 'blue', 'green', 'orange', 'purple']
```

```python
# Create a bar chart
fig, ax = plt.subplots()
ax.barh(x, y, color=stateColor)

# Add title and labels
ax.set_title('Sample Bar Chart', fontsize=16, fontweight='bold', color='navy')
ax.set_xlabel('X-axis', fontsize=10, fontweight='bold', color='navy')
ax.set_ylabel('Y-axis', fontsize=10, fontweight='bold', color='navy')

# Add data labels inside each bar
for i in range(len(x)):
    ax.text(y[i]/2, i, str(y[i]), ha='center', va='center', fontsize=9, fontweight='normal', color='white')

# Handle grid display
ax.grid(False)

# Set figure size
fig.set_size_inches(8, 6)

# Show the plot
plt.show()
```
