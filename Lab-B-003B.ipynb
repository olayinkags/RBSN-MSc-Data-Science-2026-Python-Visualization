```python
# Import necessary libraries
import matplotlib.pyplot as plt
```

```python
# Define the data
x = ['Imo', 'Abuja', "Kano", 'Lagos', "Oyo"]
y = [22, 30, 45, 74, 11]
z = [12, 20, 35, 64, 7]
```

```python
# Create a bar chart
fig, ax = plt.subplots()

# Plot the bars for y and z
ax.bar(x, y, color='orange')
ax.bar(x, z, bottom=y, color='lightblue')

# Add title and labels
ax.set_title('Sample Bar Chart', fontsize=16, fontweight='bold', color='navy')
ax.set_xlabel('X-axis', fontsize=10, fontweight='bold', color='navy')
ax.set_ylabel('Y-axis', fontsize=10, fontweight='bold', color='navy')

# Add data labels inside each bar
for i in range(len(x)):
    ax.text(i, y[i] / 2, str(y[i]), ha='center', va='center', fontsize=9, fontweight='normal', color='black')
    ax.text(i, y[i] + z[i] / 2, str(z[i]), ha='center', va='center', fontsize=9, fontweight='normal', color='black')

# Handle grid display
ax.grid(False)

# Set figure size
fig.set_size_inches(8, 6)

# Show the plot
plt.show()
```
