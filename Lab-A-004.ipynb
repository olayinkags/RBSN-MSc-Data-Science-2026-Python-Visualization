```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Define the data
data = {}
x = [1, 2, 3, 4, 5]
y = [22, 30, 45, 74, 11]

data['x'] = x
data['y'] = y

df = pd.DataFrame(data)
```

```python
# Function to handle conditional formatting
def defineColorScheme(value):
    if value < 50:
        return 'red'
    elif 50 <= value < 60:
        return 'orange'
    else:
        return 'green'
```

```python
# Apply conditional formatting to the data points
df['Color'] = df['y'].apply(defineColorScheme)
```

```python
# Create a line plot
fig, ax = plt.subplots()
ax.plot(df['x'], df['y'], color='g', linestyle='dotted', linewidth=2, marker='o', markersize=8, markerfacecolor='red', markeredgewidth=2, markeredgecolor='black')

# Add conditional coloring to the markers
for i in range(len(df)):
    ax.plot(df['x'][i], df['y'][i], marker='o', markersize=8, markerfacecolor=df['Color'][i], markeredgewidth=2, markeredgecolor='black')

    
# Add title and labels
ax.set_title('Sample Line Plot', fontsize=14, fontweight='bold', color='blue')
ax.set_xlabel('X-axis', fontsize=12, fontweight='normal', color='blue')
ax.set_ylabel('Y-axis', fontsize=12, fontweight='normal', color='blue')

ax.set_ylim(0, 100)  # Set y-axis limits
ax.set_xlim(0, 6)     # Set x-axis limits

# Handle grid display
ax.grid(False)

# Add annotation
ax.annotate('Peak Value', xy=(4, 74), xytext=(3, 80), arrowprops=dict(facecolor='black', shrink=0.05))

# Add horizontal line at y=50
ax.axhline(y=50, color='red', linestyle='--', linewidth=1)
ax.text(0.5, 50, 'Safety Line', color='red', fontsize=12, verticalalignment='center')

# Set figure size
fig.set_size_inches(8, 6)

# Show the plot
plt.show()
```
