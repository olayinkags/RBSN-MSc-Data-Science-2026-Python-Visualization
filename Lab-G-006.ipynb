```python
# Import necessary libraries
import matplotlib.pyplot as plt
```

```python
# Define the data
x = [1, 2, 3, 4, 5]
y = [22, 30, 45, 74, 11]
```

```python
# Create a line plot
fig, ax = plt.subplots()
ax.plot(x, y, color='g', linestyle='dotted', linewidth=2, marker='o', markersize=8, markerfacecolor='red', markeredgewidth=2, markeredgecolor='black')

# Color single letter options: 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'
# Color names: 'blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'white', etc.

# Add title and labels
ax.set_title('Sample Line Plot', fontsize=16, fontweight='bold', color='blue')
ax.set_xlabel('X-axis', fontsize=16, fontweight='bold', color='blue')
ax.set_ylabel('Y-axis', fontsize=16, fontweight='bold', color='blue')

# Add grid
ax.grid(True)

# Show the plot
plt.show()
```
