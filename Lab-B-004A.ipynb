```python
# Import necessary libraries
import matplotlib.pyplot as plt
```

```python
# Define the data
state = ['Imo', 'Abuja', "Kano", 'Lagos', "Oyo"]

x = [10, 20, 30, 40, 50]
y = [22, 30, 45, 74, 11]
z = [12, 20, 35, 64, 7]
```

```python
# Create a bar chart
import numpy as np

fig, ax = plt.subplots()

states = ['Imo', 'Abuja', "Kano", 'Lagos', "Oyo"]
y = [22, 30, 45, 74, 11]
z = [12, 20, 35, 64, 7]

x = np.arange(len(states))
width = 0.35

# Plot the bars for y and z side by side
ax.bar(x - width/2, y, width, color='orange', label='Y')
ax.bar(x + width/2, z, width, color='lightblue', label='Z')

# Add title and labels
ax.set_title('Sample Bar Chart', fontsize=16, fontweight='bold', color='navy')
ax.set_xlabel('States', fontsize=10, fontweight='bold', color='navy')
ax.set_ylabel('Values', fontsize=10, fontweight='bold', color='navy')

# Add data labels inside the bars
for i in range(len(states)):
    ax.text(x[i] - width/2, y[i] / 2, str(y[i]), ha='center', va='center', color='white', fontweight='normal')
    ax.text(x[i] + width/2, z[i] / 2, str(z[i]), ha='center', va='center', color='black', fontweight='normal')

# Set x-ticks and labels
ax.set_xticks(x)
ax.set_xticklabels(states)

# Add legend
ax.legend(loc='upper right')
# loc options: 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'

# Handle grid display
ax.grid(False)

# Set figure size
fig.set_size_inches(8, 6)

# Show the plot
plt.show()
```
