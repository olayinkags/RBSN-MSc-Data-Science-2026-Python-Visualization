```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

mydataset = [450, 200, 780]

# defines a pie chart
ax.pie(
    mydataset,
    startangle = 90,
    pctdistance=0.85,
    colors= ["white", "white", "green"]
    )

# defines and place a white hole at the centre of the pie chart
hole = plt.Circle((0,0), 0.70, fc="white")
plt.gcf()

fig.gca().add_artist(hole)

# defines the text overlayed on the donut
plt.text(0,0,"55%", ha="center", va="center", fontsize=50)
```
