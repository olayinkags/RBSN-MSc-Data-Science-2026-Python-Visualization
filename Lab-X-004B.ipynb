```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Labels for each bar.
categories = ['Q2', 'Sales', 'Interest', 'Fixed Costs', 'Variable Costs', 'Q3']

# Relative values. Note, the last bar for "Q3" is auto-calculated, so not given.
values = [100, 80, 20, -40, -20]
```

```python
# Calculate the cumulative sum.
cumulative_values = np.cumsum(values)
```

```python
# Create the waterfall chart
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the bars
for i in range(len(values)):
    if i == 0:
        color = 'dodgerblue'
    elif values[i] >= 0:
        color = 'green'
    else:
        color = 'firebrick'
    bar = ax.bar(
        categories[i],
        values[i],
        bottom=cumulative_values[i] - values[i],
        color=color
    )
    ax.text(
        categories[i],
        cumulative_values[i],
        f"${values[i]:,}",
        ha='center',
        va='bottom' if values[i] >= 0 else 'top',
        color='black',
        fontsize=10,
        fontweight='bold'
    )

# Plot the last bar.
final_bar = ax.bar(categories[-1], cumulative_values[-1], color='dodgerblue')
ax.text(
    categories[-1],
    cumulative_values[-1],
    f"${cumulative_values[-1]:,}",
    ha='center',
    va='bottom',
    color='black',
    fontsize=10,
    fontweight='bold'
)

# Add labels and title
ax.set_title('Financial Summary - 2024')
ax.set_ylabel('Balance ($)')

# Set x-axis labels to the middle of the bars
plt.xticks(rotation=30, ha='right')
plt.show()
```
