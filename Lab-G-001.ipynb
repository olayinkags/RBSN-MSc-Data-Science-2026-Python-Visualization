```python
# Import necessary libraries
import matplotlib.pyplot as plt
```

```python
# Define the data
x = [1, 2, 3, 4, 5]
y = [22, 30, 45, 74, 11]
```

```python
# Create a line plot
fig, ax = plt.subplots()

ax.plot(x, y)

plt.show()
```
