```python
# Import necessary libraries
import matplotlib.pyplot as plt
```

```python
# Define the data
x = [16, 12, 33, 46, 57, 28, 39, 25, 18, 42]
y = [22, 30, 45, 74, 11, 50, 60, 35, 20, 80]
```

```python
# Create a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, y, color='red', s=100, edgecolors='black', linewidth=2, alpha=0.7)

# Color single letter options: 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'
# Color names: 'blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'white', etc.

# Add title and labels
ax.set_title('Sample Scatter Plot', fontsize=16, fontweight='bold', color='blue')
ax.set_xlabel('X-axis', fontsize=10, fontweight='bold', color='blue')
ax.set_ylabel('Y-axis', fontsize=10, fontweight='bold', color='blue')

# Set axis limits
plt.xlim(0, 60)
plt.ylim(0, 80)

# Add grid
plt.grid(True)

# Show the plot
plt.show()
```
